## STEP - 1
> ### **Importing Library**

```python
import pandas as pd
```

## STEP - 2
> ### **Loading data using pandas library**

```python
data = pd.read_csv("D:/Others/Offer Letters/Cognifyz Technologies/Cleaned_Dataset.csv")
data.head(5)
```

## STEP - 3
> ### **Identify the most common combinations of cuisines**
> ### **Split the 'Cuisines' column by commas to handle multiple cuisines for a restaurant**

```python
cuisine_combinations = data['Cuisines'].value_counts().head(10)
```

## STEP - 4
> ### **Determine if certain cuisine combinations tend to have higher ratings**

```python
cuisine_ratings = data.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
```

## STEP - 5 
> ### ** Display the top 10 most common cuisine combinations and their average ratings**

```python
print(f"Most common cuisine combinations:\n{cuisine_combinations}")
print(f"\nCuisine combinations with highest ratings:\n{cuisine_ratings.head(10)}")
```
