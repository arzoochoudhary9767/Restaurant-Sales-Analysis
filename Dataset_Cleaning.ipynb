# *CLEANING OF DATASET*

## STEP - 1
>Importing Library

```python
import pandas as pd
```

## STEP - 2
>Loading of dataset using Pandas 

```python
data = pd.read_csv("D:/Others/Offer Letters/Cognifyz Technologies/Dataset .csv")
data
```

## STEP - 3 
> ### 1. Handling Missing Values
> ### Remove rows where critical columns like 'Restaurant Name', 'Cuisines', or 'City' are missing

```python
cleaned_data = data.dropna(subset=['Restaurant Name', 'Cuisines', 'City'])
```

## STEP - 4
> ### 2. Removing duplicates
> ### Remove rows with duplicate 'Restaurant ID' as it's a unique identifier

```python
cleaned_data = cleaned_data.drop_duplicates(subset='Restaurant ID')
```

## STEP - 4
> ### 3. Standardizing case
> ### Convert all strings to lower case for consistency

```python
cleaned_data['Restaurant Name'] = cleaned_data['Restaurant Name'].str.lower()
cleaned_data['City'] = cleaned_data['City'].str.lower()
cleaned_data['Cuisines'] = cleaned_data['Cuisines'].str.lower()
```

## STEP - 5
> ### 4. Handling inconsistent data in 'Has Table booking', 'Has Online delivery' (Convert Yes/No to 1/0 for easier analysis)

```python
cleaned_data['Has Table booking'] = cleaned_data['Has Table booking'].map({'Yes': 1, 'No': 0})
cleaned_data['Has Online delivery'] = cleaned_data['Has Online delivery'].map({'Yes': 1, 'No': 0})
```

## STEP - 6
> ### 5. Removing unnecessary columns
> ### If there are any columns that are not needed for the analysis, drop them

```python
cleaned_data = cleaned_data.drop(columns=['Locality Verbose', 'Switch to order menu'])
```

> ### Displaying the cleaned dataset structure

```python
cleaned_data.head()
```
