## STEP - 1

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## STEP - 2
> ### **Loading data using pandas library**

```python
data = pd.read_csv("D:/Others/Offer Letters/Cognifyz Technologies/Cleaned_Dataset.csv")
data.head(5)
```

## STEP - 3
> ### **Plot the geographic locations of restaurants using latitude and longitude**

```python
plt.figure(figsize=(10, 8))
plt.scatter(data['Longitude'], data['Latitude'], alpha=0.5, c='blue')
plt.title('Geographic Distribution of Restaurants')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
