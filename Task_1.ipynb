## Solution:
> ### **Analyze rating distribution and identify the most common rating range.**
> ### **Calculate the average votes received by restaurants.**

## STEP - 1
> ### **Importing Library**

```python
import pandas as pd
```

## STEP - 2
> ### **Loading data using pandas library**

```python
data = pd.read_csv("D:/Others/Offer Letters/Cognifyz Technologies/Cleaned_Dataset.csv")
data.head(5)
```

## STEP - 3
> ### **Analyze the distribution of aggregate ratings**

```python
rating_distribution = data['Aggregate rating'].value_counts().sort_index()
```

## STEP - 4
> ### **Find the most common rating range (grouping ratings in bins if needed, e.g., 0-1, 1-2, etc.)**

```python
rating_bins = pd.cut(data['Aggregate rating'], bins=[0, 1, 2, 3, 4, 5], right=False)
most_common_rating_range = rating_bins.value_counts().idxmax()
```

## STEP - 5
> ### **Calculate the average number of votes received by restaurants**

```python
average_votes = data['Votes'].mean()
```

## STEP - 6
> ### **Displaying the results**

```python
print(f"Rating distribution (by individual rating values):\n{rating_distribution}")
print(f"\nMost common rating range: {most_common_rating_range}")
print(f"\nAverage number of votes per restaurant: {average_votes:.2f}")
```
